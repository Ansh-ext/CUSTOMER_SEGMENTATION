```python
features=['Age','Income','Total_spending','NumWebPurchases','NumCatalogPurchases','NumStorePurchases','NumWebVisitsMonth','Recency']
```

```python
X=df[features].copy()
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
wcss=[]
silhouette_s=[]
for i in range (2,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    silhouette_s.append(silhouette_score(X_scaled,kmeans.labels_))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(range(2,10), wcss,marker='o')
ax[0].set_title("Elbow Method")

ax[1].plot(range(2,10), silhouette_s,marker='o')
ax[1].set_title("Silhouette Score")

plt.show()
```

```python
kmeans=KMeans(n_clusters=6)

df['Cluster(k_means)']=kmeans.fit_predict(X_scaled)
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch

dendrogram=sch.dendrogram(sch.linkage(X_scaled,method='ward'))
plt.show()
```

```python
model=AgglomerativeClustering(n_clusters=2,linkage='ward')
df['Cluster(heirarchical)']=model.fit_predict(X_scaled)
```

AgglomerativeClustering is suggesting 2 clusters::Level 1 (Macro segmentation)

High-value customers

Kmeans is suggesting 6(by elbow) and maybe 3-4 by silh. score::Level 2 (Micro segmentation inside each)

KMeans splits them further into:
